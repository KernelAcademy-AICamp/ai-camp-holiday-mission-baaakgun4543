## 🔍 첫 번째 데이터 확인

첫 번째 행의 `system` 컬럼 내용을 출력합니다.

- **`df.iloc[0]`**: DataFrame의 첫 번째 행(0번 인덱스) 선택
- **`['system']`**: 해당 행의 `system` 컬럼 값 가져오기
- **`system` 컬럼**: AI에게 주어지는 시스템 프롬프트 (역할 설정 + 컨텍스트 정보)

```python
import re

def extract_contexts(system_prompt):
    context_match = re.search(r'CONTEXT="""\n(.*?)"""', system_prompt, re.DOTALL)
    if not context_match:
        return []

    context_text = context_match.group(1)

    contexts = re.findall(r'\(context \d+\)=(.*?)(?=\n\(context \d+\)=|\Z)', context_text, re.DOTALL)

    cleaned_contexts = []
    for context in contexts:
        lines = context.split('\n')
        cleaned_lines = [line for line in lines if not line.strip().startswith('Title:')]
        cleaned_context = '\n'.join(cleaned_lines).strip()
        cleaned_contexts.append(cleaned_context)

    return cleaned_contexts

system_prompt = df.iloc[0]['system']

extracted_contexts = extract_contexts(system_prompt)

for i, context in enumerate(extracted_contexts, 1):
    print(f"Context {i}:")
    print(context)
    print("-" * 50)
```

## 🔧 컨텍스트 추출 함수

`system` 프롬프트에서 여러 개의 컨텍스트 문단을 추출하는 함수입니다.

### 함수 동작 원리
1. **정규표현식으로 전체 컨텍스트 블록 찾기**: `CONTEXT="""..."""` 부분 추출
2. **개별 컨텍스트 분리**: `(context 1)=...`, `(context 2)=...` 형태로 나눔
3. **Title 줄 제거**: 각 컨텍스트에서 `Title:` 로 시작하는 줄 삭제
4. **정리된 컨텍스트 리스트 반환**

### 실행 결과
- 첫 번째 데이터의 `system` 프롬프트에서 3개의 컨텍스트를 추출
- 각 컨텍스트는 금융 관련 문서 내용

```python
extracted_contexts[2] == df.iloc[0]['answer_context_summary']
```

## ✅ 데이터 검증

추출한 컨텍스트가 원본 데이터의 정답 컨텍스트와 일치하는지 확인합니다.

- **`extracted_contexts[2]`**: 추출한 컨텍스트 중 3번째 (인덱스 2)
- **`df.iloc[0]['answer_context_summary']`**: 원본 데이터의 정답 컨텍스트
- **비교 결과**: `True`면 올바르게 추출됨을 의미

```python
contexts = []
for i in range(len(df)):
    system_prompt = df.iloc[i]['system']
    extracted_contexts = extract_contexts(system_prompt)
    contexts.append(extracted_contexts)
```

## 🔄 전체 데이터 컨텍스트 추출

200개 모든 데이터에서 컨텍스트를 추출하여 리스트에 저장합니다.

- **`for i in range(len(df))`**: 모든 행을 순회 (0~199)
- **`extract_contexts()`**: 각 행의 `system` 프롬프트에서 컨텍스트 추출
- **`contexts.append()`**: 추출한 컨텍스트를 리스트에 추가
- **결과**: 200개의 질문마다 여러 개의 컨텍스트가 담긴 2차원 리스트

```python
questions = [df.iloc[i]['human'] for i in range(len(df))]
contexts_answers_idxs = [df.iloc[i]['answer_position'] - 1 for i in range(len(df))]
contexts_answers = [df.iloc[i]['answer_context_summary'] for i in range(len(df))]
answers = [df.iloc[i]['answer'] for i in range(len(df))]
```

## 📝 필요한 데이터 추출

RAG 시스템에 필요한 4가지 핵심 정보를 추출합니다.

- **`questions`**: 사용자가 물어보는 질문들 (`human` 컬럼)
- **`contexts_answers_idxs`**: 정답이 들어있는 컨텍스트의 위치 (1부터 시작 → 0부터 시작으로 변환)
- **`contexts_answers`**: 정답이 들어있는 컨텍스트 문단 (`answer_context_summary` 컬럼)
- **`answers`**: 질문에 대한 정답 (`answer` 컬럼)

각각 200개의 요소를 가진 리스트로 저장됩니다.

```python
import pickle

rag_data = {
    'questions': questions,
    'contexts': contexts,
    'contexts_answer_idx': contexts_answers_idxs,
    'contexts_answers': contexts_answers,
    'answers': answers
}

with open('./res/rag_data.pkl', 'wb') as f:
    pickle.dump(rag_data, f)
```

## 💾 데이터 저장 (Pickle)

추출한 모든 데이터를 딕셔너리로 묶어서 파일로 저장합니다.

- **`rag_data` 딕셔너리**: 5가지 핵심 정보를 key-value 형태로 저장
  - `questions`: 질문 리스트
  - `contexts`: 모든 컨텍스트 리스트 (2차원)
  - `contexts_answer_idx`: 정답 컨텍스트 위치
  - `contexts_answers`: 정답 컨텍스트 내용
  - `answers`: 정답 리스트

- **`pickle.dump()`**: 파이썬 객체를 파일로 저장 (나중에 빠르게 불러오기 가능)
- **저장 경로**: `./res/rag_data.pkl`
